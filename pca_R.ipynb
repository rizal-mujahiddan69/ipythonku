```r
library("tidyverse")
library("OpenImageR")
library(jpeg)
lar <- 1:12000
matra <- matrix(lar,1,12000,byrow = TRUE)
# lu setting pathnya sesuai dengan folder yang gw kirim deh
file_list <- list.files(path = "/content/fotopca")

for(i in file_list){
  pathi <- file.path("/content/fotopca" ,i)
  im <- readImage(pathi)
  image_r2g <- rgb_2gray(im)
  imaga <- resizeImage(image_r2g,height = 120,width = 100,method = "bilinear")
  pos <-paste(sep='',"/content/aku/",i)
  writeImage(imaga,file_name = pos)
  aku <- as.vector(imaga)
  matra <- rbind(matra,aku)
}

matra<- matra[-1,]
rownames(matra) <- NULL
matra
```

**Ini Nomor 6**

```r
covar <- cov(matra)
covar <- covar *(nrow(matra)-1)/ nrow(matra)
print("matrix covar")
covar
```

ini Nomor 7 di ketiga cell ke bawah yah

```r
eig <- eigen(covar)
eig$values
# sampai 30 menitan untuk menghitung ini deh
```

```r
dat <- eig$values
barplot(dat[1:22],  col="blue",main ="Nilai Eigen")
```

ini eigen vektornya

```r
eig$vectors
```

```r
print(dim(eig$vectors))
```

```r
akar_ciri <- eig$values
vektor_ciri <- eig$vectors
banding_proposisi <- sum(akar_ciri)
print(banding_proposisi)
```

```r
proporsi <- akar_ciri/banding_proposisi
plot(proporsi[1:22],type="h",col="red" ,ylab="proposi",main = "proposisi tiap Sumbu baru")
```

ini nomor 8

```r
g <- 0
jumlah_akar_ciri <- 0
for(i in 1:length(akar_ciri)){
  jumlah_akar_ciri <- jumlah_akar_ciri + akar_ciri[i]
  if((jumlah_akar_ciri/banding_proposisi) > 0.9){
    g<-i
    break
  }
}
paste("ini nilai optimum dari pca yah",g)
```

```r
norm_vec <- function(x){ sqrt(sum(x^2))}
for(i in 1:ncol(vektor_ciri)){
  bagi <-  norm_vec(vektor_ciri[,i])
  vektor_ciri[,i] <- vektor_ciri[,i] / bagi
 }
```
