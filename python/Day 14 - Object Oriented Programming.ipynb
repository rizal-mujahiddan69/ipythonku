### Apa itu Object?
Sebuah class adalah sebuah himpunan dari object. Sebuah Object adalah milik sebuah class. Sebuah object adalah representasi/penjelmaan dari kebutuhan, sifat, propertis dan karakteristik yang diassign ke dalam sebuah spesifik class. Artinya object adalah milik dari spesifik class(salah satu class), dan juga secara bersamaan milik dari superclass (karena menurunkan sifat dan propertinya) ke class-class yang lain.

- Setiao subclass class lebih spesifik karakteristiknya bilan dibandingkan superclass(umum)
- Sebaliknya, tiap superclass lebih abstract dibandingkan subclassnya.

Konsep **inheritance** pada dasar OOP dapat dikatakan semua sifat/karaktertik (traits) yang diwariskan kepada object di subclass berasal dari object yang didiefinisikan di awal (superclass). Jadi setiap karakteristik dan sifat object diwariskan berdasarkan pola hirarki.

```python
class Stack:
    def __init__(self):
        self.__stack_list = []

    def push(self, val):
        self.__stack_list.append(val)

    def pop(self):
        val = self.__stack_list[-1]
        del self.__stack_list[-1]
        return val


class AddingStack(Stack):
    def __init__(self):
        Stack.__init__(self)
        self.__sum = 0

    def get_sum(self):
        return self.__sum

    def push(self, val):
        self.__sum += val
        Stack.push(self, val)

    def pop(self):
        val = Stack.pop(self)
        self.__sum -= val
        return val
```

```python
stack_object = AddingStack()

for i in range(4,8):
    stack_object.push(i)
print(stack_object.get_sum())

# for i in range(5):
#     print(stack_object.pop())
```
